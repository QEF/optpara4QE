```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import collections.abc
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
import GPy
import itertools

sns.set_style("ticks")
sns.set_context("talk")

def flatten(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, collections.MutableMapping):
            items.extend(flatten(v, new_key, sep=sep).items())
        else:
            if type(v) == list:
                if type(v[1]) == str:
                    items.append((new_key, v[0]))
                    items.append((new_key + "_unit", v[1]))
                else:
                    for it, v_ in enumerate(v):
                        items.append((new_key + f"_{it}", v_))

            else:
                items.append((new_key, v))
    return dict(items)

def train_test_split(df_, split=0.8):
    # Create training and test set
    msk = np.random.rand(len(df_)) < split
    df_train = df_[msk]
    df_val = df_[~msk]

    X_tr = df_train.drop([columns_to_keep[0]], axis=1).values
    Y_tr = df_train[columns_to_keep[0]].values

    X_val = df_val.drop([columns_to_keep[0]], axis=1).values
    Y_val = df_val[columns_to_keep[0]].values
    
    return X_tr, Y_tr, X_val, Y_val

def denorm(y):
    return np.exp(y-1)

def norm(y):
    return np.log(y)+1
```

# Data Gathering

```python
# Load data into dictionary
with open("qnp.json") as f :
    data = json.load(f)
    
cname = '$\dfrac{N_{el}^3 \cdot N_{k}}{N_{cores}}$'
tname = '$\dfrac{t_{e}}{N_{iter}^{scf}}$ [s]'

# Flatten all nested dictionaries
data = [flatten(d) for d in data]
# Transform into dataframe, dropping the first row (outlier) and renaming the most important time column
df = pd.DataFrame.from_dict(data).drop(0)
# df = df.rename(columns = {'clocks_PWSCF': 'time'})

df = df.rename(columns = {'dims_Threads': 'Threads per node'})
```

```python
# User definible variables: dims_Threads/dims_MPI tasks, dims_npool, dims_ndiag
# System variables: dims_nkpoints, dims_nat, n_resources
# Non controllable but meaningful variables: iter_sum_band
# in this case dims_nkpoints and dims_nat always vary together, so I keep only dims_nat
# With fixed resources, dims_Threads and dims_MPI tasks are inveresely proportional, so I only keep dims_Threads

# dims_ndiag: how resources for scalapack are allocated
# dims_npool: how cores are divided for k points
# dims_Threads*dims_MPI tasks is constant (OpenMP vs MPI parallelization)

# Other columns that might be of interest are dims_wfc_fft_division_0 and dims_wfc_fft_division_0; which product is always constant within a setup
```

```python
# Remove rows that only contain one unique value and transform all to float
df_clean = pd.DataFrame()
for c in list(df.columns):
    if len(df[c].unique()) > 1:
        try:
            df_clean[c] = df[c].astype('float')
        except ValueError:
            pass

# The column we use as time is the time per call of the iteration
df_clean[tname] = df_clean['clocks_PWSCF']*(1/df_clean['iter_sum_band'])

df_clean['n_cores'] = df_clean['Threads per node']*df_clean['dims_MPI tasks']//48
df_clean['nel^3'] = (df_clean['dims_nat']*df_clean['eperatom'])**3
df_clean[cname] = df_clean['nel^3']/df_clean['n_cores']*df_clean['dims_nkpoints']

df_clean['Norm. Time Per Call'] = df_clean[tname]/df_clean[cname]*1000000
df_clean = df_clean.drop(index=738) # Remove an outlier

# The first column is alwyas the target
columns_to_keep = ['Norm. Time Per Call',  'nel^3',  'n_cores', 'Threads per node', 'dims_npool']
df_x = pd.DataFrame()
for c in columns_to_keep:
    df_x[c] = df_clean[c]
```

# Create training and test set

```python
df_norm = norm(df_x)
cross_vals = 5
```

# Linear Regression

```python
Y_vals, Y_hats = [], []
for r in np.arange(cross_vals):
    X_tr, Y_tr, X_val, Y_val = train_test_split(df_norm)
    reg = LinearRegression().fit(X_tr, Y_tr)
    Y_hat = reg.predict(X_val)
    Y_vals.extend(Y_val)
    Y_hats.extend(Y_hat)
Y_hats = np.array(Y_hats)
Y_vals = np.array(Y_vals)

plt.scatter(denorm(Y_val), denorm(Y_hat))
plt.plot(denorm(Y_val), denorm(Y_val), 'k-')
rmse = mean_squared_error(denorm(Y_val), denorm(Y_hat))**0.5
plt.xlabel("True %s" %(columns_to_keep[0]))
plt.ylabel("Predicted %s" %(columns_to_keep[0]))
plt.text(0, 8e-5, "RMSE: %.2f" %(rmse))
plt.show()
```

```python
ordering = np.argsort(abs(reg.coef_))[::-1]
plt.bar(df_x.drop([columns_to_keep[0]], axis=1).columns[ordering], reg.coef_[ordering])
plt.xticks(rotation='vertical', ticks = np.arange(len(columns_to_keep[1:])), labels = columns_to_keep[1:])
plt.ylabel("Parameter Weight")
plt.show()
```

# LASSO

```python
Y_vals, Y_hats = [], []
for r in np.arange(cross_vals):
    X_tr, Y_tr, X_val, Y_val = train_test_split(df_norm)
    clf = Lasso(alpha=.05).fit(X_tr, Y_tr)
    Y_hat = clf.predict(X_val)
    Y_vals.extend(Y_val)
    Y_hats.extend(Y_hat)
Y_hats = np.array(Y_hats)
Y_vals = np.array(Y_vals)

plt.scatter(denorm(Y_vals), denorm(Y_hats))
plt.plot(denorm(Y_vals), denorm(Y_vals), 'k-')
rmse = mean_squared_error(denorm(Y_vals), denorm(Y_hats))**0.5
plt.xlabel("True %s" %(columns_to_keep[0]))
plt.ylabel("Predicted %s" %(columns_to_keep[0]))
plt.text(0, 5e-5, "RMSE: %.2f s" %(rmse))
plt.show()
```

```python
ordering = np.argsort(abs(clf.coef_))[::-1]
plt.bar(df_x.drop([columns_to_keep[0]], axis=1).columns[ordering], reg.coef_[ordering])
plt.xticks(rotation='vertical')
plt.ylabel("Parameter Weight")
plt.show()
```

# Gaussian Process Regression

```python
# Create kernel with 1 independent RBF per component
k1  = GPy.kern.RBF(input_dim=1, active_dims = [0])
k2  = GPy.kern.RBF(input_dim=1, active_dims = [1])
k3  = GPy.kern.RBF(input_dim=1, active_dims = [2])
k4  = GPy.kern.RBF(input_dim=1, active_dims = [3])
k5  = GPy.kern.RBF(input_dim=1, active_dims = [4])
k6  = GPy.kern.RBF(input_dim=1, active_dims = [5])
k7  = GPy.kern.RBF(input_dim=1, active_dims = [6])
k8  = GPy.kern.RBF(input_dim=1, active_dims = [7])


k = k1*k2*k3*k4#*k5*k6#*k7*k8#*k9*k10*k11*k12*k13*k14*k15

Y_vals, Y_hats = [], []
for r in np.arange(cross_vals):
    print(r)
    X_tr, Y_tr, X_val, Y_val = train_test_split(df_norm)
    # Optimize hyperparameters automatically and fit
    k = k1*k2*k3*k4#*k5*k6#*k7*k8
    m = GPy.models.GPRegression(X_tr, Y_tr.reshape(len(Y_tr), 1), kernel = k, noise_var = 0.4)
    opts = m.optimize(messages=False)
    Y_hat = m.predict(X_val)[0]
    Y_vals.extend(Y_val)
    Y_hats.extend(Y_hat)
Y_hats = np.array(Y_hats)
Y_vals = np.array(Y_vals)
```

```python
plt.scatter(denorm(Y_vals), denorm(Y_hats))
plt.plot(denorm(Y_vals), denorm(Y_vals), 'k-')
rmse = mean_squared_error(denorm(Y_vals), denorm(Y_hats))**0.5
plt.xlabel("True %s" %(columns_to_keep[0]))
plt.ylabel("Predicted %s" %(columns_to_keep[0]))
plt.text(0, 6e-5, "RMSE: %.2f" %(rmse))
plt.show()
```

```python
def get_optimal_parallelization(m, nel, ncores):
    yopt = 1e6
    threads_per_node = np.array([2, 4, 8])
    npool = np.array([1, 2, 4, 8])
    x_ = np.array([a for a in itertools.product(threads_per_node, npool)])
    x_ = np.hstack([nel*np.ones((len(x_), 1)), ncores*np.ones((len(x_), 1)), x_])
    x_ = x_[x_[:, 3] <= x_[:, 1]]
    y = denorm(m.predict(norm(x_))[0])
    yopt = np.min(y)
    xopt = np.round(x_[np.argmin(y)])
    return xopt, yopt


yopt_pred = []
yopt_true = []
xopt_pred = []
xopt_true = []
yopt_xopt = []
ymean = []

for x in X_val:
    yopt = 1e6
    nel = denorm(x)[0]
    ncores = denorm(x)[1]
    xopt, yopt = get_optimal_parallelization(m, nel, ncores)
    xopt_pred.append(xopt)
    yopt_pred.append(yopt)
    yopt_xopt.append(df_x.loc[(df_x['nel^3'] == xopt[0]) & (df_x.n_cores == xopt[1]) & 
                              (df_x['Threads per node'] == xopt[2]) & (df_x.dims_npool == xopt[3])].mean()['Norm. Time Per Call'])
    yopt_true_ = df_x[(df_x['nel^3'] == xopt[0]) & (df_x.n_cores == xopt[1])].min()['Norm. Time Per Call']
    yopt_true.append(yopt_true_)
    xopt_true.append(df_x[df_x['Norm. Time Per Call'] == yopt_true_].values[0, 1:])
    
    ymean.append(df_x[(df_x['nel^3'] == xopt[0]) & (df_x.n_cores == xopt[1])].mean()['Norm. Time Per Call'])
    
yopt_pred = np.array(yopt_pred)
yopt_true = np.array(yopt_true)
xopt_pred = np.array(xopt_pred)
xopt_true = np.array(xopt_true)
yopt_xopt = np.array(yopt_xopt)
ymean     = np.array(ymean)
```

```python
plt.scatter(yopt_true, yopt_xopt)
plt.plot(yopt_true, yopt_true, 'k-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Optimal Time')
plt.ylabel('Time For Suggested Setup')
plt.show()

plt.scatter(yopt_true, ymean)
plt.plot(yopt_true, yopt_true, 'k-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Optimal Time')
plt.ylabel('Average Time For Problem')
plt.show()

plt.scatter(yopt_xopt, ymean)
plt.plot(yopt_xopt, yopt_xopt, 'k-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Time For Suggested Setup')
plt.ylabel('Average Time For Problem')
plt.show()
```
